<h1>Using Weighted Average</h1>

formula ==> W =Rv+Cm/v+m

<p1>W ---> Weighted rate
R----> avg for the movie number from 0to10
C----> the mean vote across the report
v----> no.of votes for movie
m----> min votes required to be listed movies.</p1>

```python
v = moives_merge['vote_count']
R = moives_merge['vote_average']
C = moives_merge['vote_average'].mean()
m = moives_merge['vote_count'].quantile(0.70)
```

```python
moives_merge['Weighted_Average']=((R*v)+(C*m)/(v+m))
```

```python
moives_merge.head()
```

```python
moives_ranking=moives_merge.sort_values('Weighted_Average',ascending=False)
moives_ranking=moives_ranking[['original_title','vote_count','vote_average','Weighted_Average','popularity']]
```

```python
moives_ranking.head(10)
```

```python
sns.barplot(x=moives_ranking['Weighted_Average'].head(10),y=moives_ranking['original_title'].head(10))
```

```python
sns.barplot(x=moives_ranking['popularity'].head(10),y=moives_ranking['original_title'].head(10))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling =MinMaxScaler()
movie_scaled = scaling.fit_transform(moives_merge[['Weighted_Average','popularity']])
movie_norm = pd.DataFrame(movie_scaled,columns=['Weighted_Average','popularity'])
```

```python
movie_norm.head()
```

```python
moives_merge[['norm_Weighted_Average','norm_popularity']]=movie_norm
```

```python
moives_merge.head()
```

```python
moives_merge['score']=moives_merge['norm_Weighted_Average']*0.5+moives_merge['norm_popularity']*0.5
```

```python
movie_score=moives_merge.sort_values(['score'],ascending=False)
movie_score=movie_score[['original_title','vote_count','vote_average','norm_Weighted_Average','norm_popularity']]
```
